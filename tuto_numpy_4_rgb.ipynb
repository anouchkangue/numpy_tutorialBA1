```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

# Lire une image depuis un fichier

```python
rgb = plt.imread('green.png')
rgb.shape
```

# afficher une image

```python
plt.imshow(rgb)
```

# composantes r,g et b

```python
plt.figure(figsize=[10,10])
plt.subplot(2,2,1)
plt.imshow(rgb)
plt.xlabel('rgb')

plt.subplot(2,2,2)
plt.imshow(rgb[:,:,0])
plt.xlabel('r')

plt.subplot(2,2,3)
plt.imshow(rgb[:,:,1])
plt.xlabel('g')

plt.subplot(2,2,4)
plt.imshow(rgb[:,:,2])
plt.xlabel('b')
```

```python
r = rgb[:,:,0]
plt.imshow(r)
plt.colorbar();
```

```python
R = r*255
```

# histogramme d'une image

```python

h,bins = np.histogram(R.flatten(),bins=range(256))
```
